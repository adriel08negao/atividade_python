```python
#Exercício 1

def contar_pares(*n):
    contador = 0
    for numero in n:
        if numero % 2 == 0:
            contador += 1
    return contador
```

```python
#Exercício 2

def cubica(*nums):
    resultado = []
    for n in nums:
        resultado.append(n ** 3)
    return resultado

print(cubica(3,6,9,12))
```

```python
#Exercício 3

def lista_compras(**produtosValor):
    total = 0
    for produto, valor in produtosValor.itens():
        print(f"{produto}: R${valor:.2f}")
        total += valor
    print(f"Total: R${total:.2f}")

    lista_compras(arroz=10, feijao=8, leite=5)
```
