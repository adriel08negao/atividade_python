```python
#Exercício 3

def converter_temperatura(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    kelvin = celsius + 273.15

    print("Celsius:", celsius)
    print("Kelvin:", kelvin)

converter_temperatura(68)
```
